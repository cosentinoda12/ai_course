```
import numpy as np
```

```
fig = plt.figure()

for i in range(10):
    digit = np.mean(train_images[train_labels==i], axis=0)
    fig.add_subplot(2, 5, i+1)
    plt.imshow(digit)
    
for axi in fig.axes:
    axi.axis("off")
plt.show()
```
