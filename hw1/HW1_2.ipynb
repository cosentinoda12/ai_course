```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
from mpl_toolkits.mplot3d import Axes3D
```

```
import numpy as np
np.random.seed(42)

xs = 2 * np.random.rand(100, 2)
ys = 4 + np.sum(xs, axis=1).reshape((-1,1)) + np.random.rand(100, 1)
```

```
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(xs[:,0], xs[:,1], ys)
```

```
# split the data into training and test sets
# train set
train_xs = xs[:80]
train_ys = ys[:80]
# test set
test_xs = xs[80:]
test_ys = ys[80:]
```

# Linear regression using numpy


```
# number of epochs
epochs = 20
# learning rate
lr = 0.01
```

```
# initial value for weight w and bias b
w = np.random.randn(2)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = np.sum(w * train_xs[i]) + b

    grad_w = (y_pred - train_ys[i]) * train_xs[i]
    grad_b = (y_pred - train_ys[i])
    w -= lr * grad_w
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (np.sum(w * test_xs[i]) + b - test_ys[i]) ** 2
test_loss /= 20
```

```
test_loss
```

```
pred_ys = np.sum(w * test_xs, axis=1) + b
```

```
plt3d = plt.figure().gca(projection='3d')
ax = plt.gca()

ax.scatter(test_xs[:,0], test_xs[:,1], test_ys, color='blue')
ax.scatter(test_xs[:,0], test_xs[:,1], pred_ys, color='red')
```

```
plt.plot(test_xs, test_ys, "b.")
plt.plot(test_xs, pred_ys, "r.") # predicted values
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```
