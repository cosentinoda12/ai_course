```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

# Function #1

```
def f1(x,y):
  return -x**2-y**2
```

```
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```
# Code from three_dimensional_plotting notebook
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.scatter3D([0],[0], [0], color='red')

ax.scatter3D([0], [-6], [-36], color='blue')
ax.scatter3D([-6], [0], [-36], color='blue')
ax.scatter3D([0], [6], [-36], color='blue')
ax.scatter3D([6], [0], [-36], color='blue')
```

```
ax.view_init(60, 35)
fig
```

```
# Quiver of Gradient
```

```
X = np.arange(-10, 10, 1)
Y = np.arange(-10, 10, 1)
X, Y = np.meshgrid(x, y)
U = -2*X
V = -2*Y

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

plt.show()
```
